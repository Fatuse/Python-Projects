```python
#defining rp35
rp35__0014 = 0.014
rp35__0025= 0.025
rp35__004 = 0.04
rp35__01 = 0.1
rp35__02 = 0.2
rp35__05 = 0.5
rp35_1 = 1
rp35_2 = 2
rp35_4 = 4
rp3S_10 = 10
rp35_20 = 20
```

```python
def perm_rp35 (phi, rp35):
    perm_rp35 = phi * ((rp35/2.665)** (1/0.45))
    return perm_rp35
```

```python
porosity = np.linspace(0.1, 30, 120)
porosity
```

```python
df_final = pd.DataFrame(data = porosity, columns = ['porosity'])
df_final
```

```python
df_final['perm_rp35_20'] = perm_rp35(df_final['porosity'], rp35_20)
df_final['perm_rp35_10'] = perm_rp35(df_final['porosity'], rp3S_10)
df_final['perm_rp35_4'] = perm_rp35(df_final['porosity'], rp35_4)
df_final['perm_rp35_2'] = perm_rp35(df_final['porosity'], rp35_2)
df_final['perm_rp35_1'] = perm_rp35(df_final['porosity'], rp35_1)
df_final['perm_rp35__05'] = perm_rp35(df_final['porosity'], rp35__05)
df_final['perm_rp35__02'] = perm_rp35(df_final['porosity'], rp35__02)
df_final['perm_rp35__01'] = perm_rp35(df_final['porosity'], rp35__01)
df_final['perm_rp35__004'] = perm_rp35(df_final['porosity'], rp35__004)
df_final['perm_rp35__0025'] = perm_rp35(df_final['porosity'], rp35__0025)
df_final['perm_rp35__0014'] = perm_rp35(df_final['porosity'], rp35__0014)
df_final
```
